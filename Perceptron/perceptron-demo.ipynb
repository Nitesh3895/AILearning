```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(x=df['cgpa'],y=df['resume_score'],hue=df['placed'])
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
X
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```

```python
# Create a linear regression to fit X, and Y
from sklearn.linear_model import LinearRegression

# Assuming X and Y are defined and hold your data
# X should be a 2D array-like object and Y should be 1D array-like object

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)
```

```python
# Display the weights and bias
print(model.coef_, model.intercept_)
```

```python
# Draw the decision boundary
plot_decision_regions(X.values, y.values, clf=model, legend=2)
```
